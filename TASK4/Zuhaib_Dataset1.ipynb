```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
# Loading the dataset
group = pd.read_csv("group-marks.csv")

# exploration
print("First 5 rows:")
display(group.head())

print("\nLast 5 rows:")
display(group.tail())

print("\nInfo:")
group.info()

print("\nSummary statistics:")
display(group.describe())
```

```python
# Check missing values
print(group.isnull().sum())
```

```python
group['math'] = pd.to_numeric(group['math'], errors='coerce')
group
```

```python
group['math'] = group['math'].fillna(group['math'].mean())
group['english'] = group['english'].fillna(group['english'].mean())
group['group'] = group['group'].fillna(group['group'].mode()[0])
group

```

```python
# Check missing values again
print(group.isnull().sum())
```

```python
group.drop_duplicates(inplace=True)
```

```python
group['rollno'] = group['rollno'].astype(str)
group['age'] = group['age'].astype(int)
group['scholarship'] = group['scholarship'].astype(int)
group
```

```python
group.rename(columns={
    'rollno': 'Student_ID',
    'name': 'Name',
    'gender': 'Gender',
    'group': 'Group_Name',
    'session': 'Session',
    'age': 'Age',
    'scholarship': 'Scholarship',
    'math': 'Math_Marks',
    'english': 'English_Marks',
    'urdu': 'Urdu_Marks'
}, inplace=True)
group
```

```python
group['Total_Marks'] = group[['Math_Marks', 'English_Marks', 'Urdu_Marks']].sum(axis=1)
group['Average_Marks'] = group['Total_Marks'] / 3
```

```python
def assign_grade(x):
    if x >= 80:
        return 'A'
    elif x >= 60:
        return 'B'
    elif x >= 40:
        return 'C'
    else:
        return 'F'

group['Grade'] = group['Average_Marks'].apply(assign_grade)
```

```python
le = LabelEncoder()
group['Gender_Encoded'] = le.fit_transform(group['Gender'])
group['Group_Encoded'] = le.fit_transform(group['Group_Name'])
group['Grade_Encoded'] = le.fit_transform(group['Grade'])
group
```

```python
scaler = MinMaxScaler()
group[['Scaled_Age', 'Scaled_Average_Marks']] = scaler.fit_transform(group[['Age', 'Average_Marks']])
group
```

```python
print("Cleaned Dataset Info:\n")
print(group.info())
print("\n Sample of Cleaned Dataset:\n")
display(group.head())
```
