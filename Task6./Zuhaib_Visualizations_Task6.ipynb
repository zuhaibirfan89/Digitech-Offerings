```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  
```

```python
df = pd.read_csv("fifa.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
df['Nationality'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Nationalities")
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Overall'].head(200))  # first 200 players
plt.title("Player Overall Rating Trend (First 200 Players)")
plt.xlabel("Player Index")
plt.ylabel("Overall Rating")
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['Preferred Foot'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Preferred Foot Distribution")
plt.ylabel("")
plt.show()
```

```python
numeric_features = [
    'Overall', 'Potential', 'Acceleration', 'SprintSpeed', 
    'Stamina', 'Strength', 'Dribbling', 'ShortPassing', 'Finishing',
    'BallControl', 'ShotPower'
]

plt.figure(figsize=(12,8))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap of Key Player Attributes")
plt.show()
```

```python
sns.pairplot(df[['Overall', 'Potential', 'Dribbling', 'Finishing']])
plt.show()
```

```python
plt.figure(figsize=(12,5))
top_positions = df['Position'].value_counts().head(10)
sns.countplot(x=df['Position'], order=top_positions.index)
plt.title("Top 10 Most Common Player Positions")
plt.xlabel("Position")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.histplot(df['Stamina'], ax=axes[0], kde=True)
axes[0].set_title("Stamina Distribution")

sns.histplot(df['Strength'], ax=axes[1], kde=True)
axes[1].set_title("Strength Distribution")

plt.tight_layout()
plt.show()
```

# SUMMARY

```python
#Dataset
I used the FIFA Dataset (Kaggle), containing 18K+ football players with 89 features including skill ratings, club info, physical attributes, and contract data.

# Key Visuals 
Bar Chart – Nationalities: Shows which countries produce the most players.
Line Chart – Overall Ratings: Helps visualize variations in rating levels.
Pie Chart – Preferred Foot: Reveals proportion of left vs right-footed players.
Heatmap: Shows strong correlations (e.g., Dribbling ↔ BallControl).
Pairplot: Highlights how skills relate (e.g., Overall ↔ Potential).
Countplot – Positions: Displays the most common playing positions.

# Insights
Most players are right footed.
Top nationalities include England, Germany, Spain, Argentina, Brazil**.
Strong correlations found:
BallControl ↔ Dribbling
Finishing ↔ ShotPower
Overall ↔ Potential
Player positions such as ST, GK, CB, CMappear most frequently.
Skills show clustering: attacking players have higher finishing/dribbling.

Overall, visualizations make it easier to understand performance patterns and player distributions across key FIFA attributes.
```
