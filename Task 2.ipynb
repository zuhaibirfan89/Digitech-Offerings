```python
query = """
SELECT 
    t.Name AS TrackName,
    SUM(il.Quantity) AS TotalSold,
    ROUND(SUM(il.UnitPrice * il.Quantity), 2) AS TotalRevenue
FROM InvoiceLine il
JOIN Track t ON il.TrackId = t.TrackId
GROUP BY t.Name
ORDER BY TotalRevenue DESC
LIMIT 10;
"""

cursor.execute(query)
result = cursor.fetchall()

print("Top 10 Selling Tracks:")
for row in result:
    print(f"Track: {row[0]}, Sold: {row[1]}, Revenue: ${row[2]}")
```

```python
query = """
SELECT 
    i.BillingCountry AS Country,
    ROUND(SUM(il.UnitPrice * il.Quantity), 2) AS TotalRevenue
FROM InvoiceLine il
JOIN Invoice i ON il.InvoiceId = i.InvoiceId
GROUP BY i.BillingCountry
ORDER BY TotalRevenue DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

print(" Revenue by Country:")
for row in result:
    print(f"{row[0]} : ${row[1]}")
```

```python
query = """
SELECT 
    STRFTIME('%Y-%m', i.InvoiceDate) AS Month,
    ROUND(SUM(il.UnitPrice * il.Quantity), 2) AS MonthlyRevenue
FROM InvoiceLine il
JOIN Invoice i ON il.InvoiceId = i.InvoiceId
GROUP BY Month
ORDER BY Month;
"""

cursor.execute(query)
result = cursor.fetchall()

print(" Monthly Revenue Performance:")
for row in result:
    print(f"{row[0]} : ${row[1]}")
```

```python
query = """
SELECT 
    t.Name AS TrackName,
    ROUND(SUM(il.Quantity * il.UnitPrice), 2) AS Revenue,
    RANK() OVER (ORDER BY SUM(il.Quantity * il.UnitPrice) DESC) AS Rank
FROM InvoiceLine il
JOIN Track t ON il.TrackId = t.TrackId
GROUP BY t.Name
ORDER BY Rank
LIMIT 10;
"""

cursor.execute(query)
result = cursor.fetchall()

print("Top Selling Tracks (Ranked):")
for row in result:
    print(f"Rank {row[2]} | Track: {row[0]} | Revenue: ${row[1]}")
```

```python
cursor.close()
sqliteConnection.close()
print("\nConnection closed successfully.")
```
