```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic
```

```python
print("Number of rows:", titanic.shape[0])
print("Number of columns:", titanic.shape[1])
```

```python
display(titanic.head())
```

```python
display(titanic.info())
```

```python
display(titanic.describe())
```

```python
print(titanic.isnull().sum())
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Cabin'] = titanic['Cabin'].fillna("Unknown")
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
print(titanic.isnull().sum())
```

```python
print(len(titanic))
```

```python
titanic = titanic.drop_duplicates()
print(len(titanic))
```

```python
titanic['Age'] = titanic['Age'].astype(float)
titanic['Fare'] = titanic['Fare'].astype(float)
```

```python
titanic = titanic.rename(columns={'Sex': 'Gender', 'Pclass': 'Passenger_Class'})
```

```python
titanic['Gender'] = titanic['Gender'].map({'male': 0, 'female': 1})
titanic['Embarked'] = titanic['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

titanic[['Age', 'Fare']] = titanic[['Age', 'Fare']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

titanic['Family_Size'] = titanic.apply(lambda row: row['SibSp'] + row['Parch'] + 1, axis=1)
print(titanic)
```

```python
# Shape of dataset after all the cleaning
print("Number of rows:", titanic.shape[0])
print("Number of columns:", titanic.shape[1])
```
